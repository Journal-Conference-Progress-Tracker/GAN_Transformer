```python
import pandas as pd
import matplotlib.pyplot as plt

# Read your summary DataFrame
summary_df = pd.read_csv('./augmentation_performance_sizes.csv')

# Filter by a specific Generation Size (e.g., 1000) to focus the comparison
df_filtered = summary_df[summary_df['Generation Size'] == 10000].copy().drop("Generation Size", axis=1)

# Set "Train Samples" as the index for clearer plotting
df_filtered.set_index('Train Samples', inplace=True)


# Create a grouped bar chart
ax = df_filtered.plot(kind='bar', figsize=(10, 6))

# (Optional) Narrow the y-axis range to emphasize differences
plt.ylim(0.75)  # Adjust as needed

# Add titles and labels
plt.title("Comparison of Accuracies (Generation Size=1000)")
plt.xlabel("Train Samples")
plt.ylabel("Accuracy")
plt.legend(title="Method")

# (Optional) Add data labels on each bar to see exact values
for container in ax.containers:
    ax.bar_label(container, fmt="%.3f", label_type='center', rotation=90)

plt.tight_layout()
plt.show()
```
